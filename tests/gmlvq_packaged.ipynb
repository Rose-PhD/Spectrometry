```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn_lvq import GmlvqModel
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

## Sample data loads

```python
def load_exp_data(_load_fn):
    x, y = _load_fn(return_X_y= True)
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)
    return x_train, x_test, y_train, y_test


x_train, x_test, y_train, y_test = load_exp_data(load_breast_cancer)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape, sep='|')

model = GmlvqModel(prototypes_per_class=1, random_state=42)
model.fit(x_train, y_train)
    
y_pred = model.predict(x_test)
print(y_pred.shape)
```
